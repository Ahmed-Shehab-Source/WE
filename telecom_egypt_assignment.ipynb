## Question 3 — Calculate a Phone Bill

Write a function called `calculate_bill(minutes, rate)` that calculates
the total phone bill based on the following rules:

Rules:
- Total = minutes × rate per minute
- Minimum charge is 15 EGP (if total is less than 15, charge 15 EGP)
- If minutes used are more than 100, apply a 20% discount on the total

Parameters:
- minutes : number of minutes used (**float**)
- rate    : price per minute in EGP (**float**), default = 0.50

Returns:
- The final bill amount (float)

Then ask the user to enter the number of minutes,
call the function, and print the result.

###**Expected Output Example:**
-----------------------------
```
Minutes Used : 120
Rate         : 0.50 EGP/min
Discount     : 20% applied
Total Bill   : 48.00 EGP
```
-----------------------------
```
Minutes Used : 30
Rate         : 0.50 EGP/min
Discount     : None
Total Bill   : 15.00 EGP  ← minimum charge applied
```
-----------------------------


````python
def calculate_bill(minutes, rate=0.50):
    total = minutes * rate
    if minutes > 100:
        total = total * 0.80
    if total < 15:
        total = 15
    return total
minutes = float(input("Enter the number of minutes used: "))
rate = 0.50
bill = calculate_bill(minutes, rate)

print("Minutes Used :", minutes)
print("Rate         :", rate, "EGP/min")

if minutes > 100:
    print("Discount     : 20% applied")
else:
    print("Discount     : No discount")

print("Total Bill   :", round(bill, 2), "EGP")
````

## Question 5 — Analyze Data Usage

Write a **function** called `analyze_usage` that takes a list of data usage values (in GB)  
and returns a dictionary containing:
- `average` — the average usage
- `max` — the highest usage
- `min` — the lowest usage

Then call the function with this list and print the results:
```python
usage = [4.5, 12.0, 7.3, 9.8, 2.1, 15.0, 6.6]
```

**Expected Output Example:**
```
Average Usage : 8.19 GB
Max Usage     : 15.0 GB
Min Usage     : 2.1 GB
```



````python
def analyze_usage(data):
    result = {}
    result["average"] = sum(data) / len(data)
    result["max"] = max(data)
    result["min"] = min(data)
    return result


usage = [4.5, 12.0, 7.3, 9.8, 2.1, 15.0, 6.6]

result = analyze_usage(usage)

print("Average Usage :", round(result["average"], 2), "GB")
print("Max Usage     :", result["max"], "GB")
print("Min Usage     :", result["min"], "GB")
````

## Question 6 — Save Complaints to a File

Write a program with two functions:

1. `add_complaint(phone, complaint)` — saves the complaint to a file called `complaints.txt`  
   in the format: `phone | complaint`

2. `show_complaints()` — reads and prints all saved complaints from the file.  
   If the file does not exist, print: `No complaints found.`

Ask the user for their phone number and complaint, then call both functions.

**Expected Output Example:**
```
Complaint saved successfully!

--- All Complaints ---
01512111111 | Internet is very slow
01501000567 | Cannot make calls
```


````python
def add_complaint(phone, complaint):
    file = open("complaints.txt", "a")
    file.write(phone + " | " + complaint + "\n")
    file.close()
    print("Complaint saved successfully!")
def show_complaints():
    try:
        file = open("complaints.txt", "r")
        print("\n--- All Complaints ---")
        print(file.read())
        file.close()
    except FileNotFoundError:
        print("No complaints found.")

phone = input("Enter phone number: ")
complaint = input("Enter complaint: ")

add_complaint(phone, complaint)
show_complaints()
````

### Part A — Base Class

Create a class called `Subscriber` with:

**Attributes (`__init__`):**
- `name`, `phone`, `package` (Bronze / Silver / Gold), `balance` (default = 0)

**Package Prices (store inside the class):**
| Package | Monthly Price |
|---------|--------------|
| Bronze  | 50 EGP       |
| Silver  | 100 EGP      |
| Gold    | 200 EGP      |

**Methods:**
- `recharge(amount)` — adds amount to balance, prints new balance.  
  Raise `ValueError` if amount ≤ 0.
- `deduct(amount)` — deducts amount from balance.  
  Raise `ValueError` if amount ≤ 0.  
  Raise `Exception` if balance is insufficient.
- `upgrade_package(new_package)` — changes the package **only if**:
  1. `new_package` is valid (Bronze / Silver / Gold)
  2. Balance is enough to cover the new package price  
  Deduct the price and confirm. If not enough → print how much is needed.
- `show_info()` — prints all subscriber details including balance.
- `__str__()` — returns a formatted string of the subscriber (used when printing the object directly).

### Part B

Create two subclasses:

**`PrepaidSubscriber(Subscriber)`:**

**`PostpaidSubscriber(Subscriber)`:**
- Has an extra attribute: `credit_limit` (default = 500 EGP)
- Override `deduct(amount)`: allow balance to go **negative** but not below `-credit_limit`.  
  If it exceeds the limit → raise `Exception("Credit limit reached!")`.
- Add method `show_debt()`: if balance is negative, print how much the subscriber owes.

### Part D

Using the list from Part C:

```
**Expected Output Example:**
--- All Subscribers ---

[Postpaid] Karim Sami  | 01525556789 | Gold   | Balance: 0.00 EGP  | Limit: 500 EGP

[Postpaid] Sara Khaled | 01534555090 | Bronze | Balance: 0.00 EGP  | Limit: 500 EGP

[Base]     Omar Mahmoud| 01645678901 | Silver | Balance: 0.00 EGP
Recharging Ahmed Ali with 100 EGP...

✗ Credit limit reached!

Top Balance: Omar Mahmoud | 50.00 EGP

```

````python
# Q7

class Subscriber:
    prices = {
        "Bronze": 50,
        "Silver": 100,
        "Gold": 200
    }
    def __init__(self, name, phone, package, balance=0):
        self.name = name
        self.phone = phone
        self.package = package
        self.balance = balance

    def recharge(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")

        self.balance = self.balance + amount
        print("New Balance:", self.balance, "EGP")

    def deduct(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")

        if amount > self.balance:
            raise Exception("insufficient balance.")

        self.balance = self.balance - amount

    def upgrade_package(self, new_package):
        if new_package not in Subscriber.prices:
            raise ValueError("invalid package.")

        price = Subscriber.prices[new_package]

        if self.balance >= price:
            self.balance = self.balance - price
            self.package = new_package
            print("Package updated successfully!")
        else:
            print("Need", price - self.balance, "EGP more.")

    def show_info(self):
        print("Name:", self.name)
        print("Phone:", self.phone)
        print("Package:", self.package)
        print("Balance:", self.balance, "EGP")

    def __str__(self):
        return self.name + " | " + self.phone + " | " + self.package + " | Balance: " + str(self.balance) + " EGP"


class PrepaidSubscriber(Subscriber):

    def __init__(self, name, phone, package, balance=0):
        super().__init__(name, phone, package, balance)
        self.data_gb = 0

    def recharge(self, amount):
        super().recharge(amount)
        self.data_gb = self.data_gb + (amount // 10)
        print("Data:", self.data_gb, "GB")

    def check_data(self):
        print("Remaining Data:", self.data_gb, "GB")

    def __str__(self):
        return "[Prepaid] " + self.name + " | " + self.phone + " | " + self.package + " | Balance: " + str(self.balance) + " EGP | Data: " + str(self.data_gb) + " GB"


class PostpaidSubscriber(Subscriber):

    def __init__(self, name, phone, package, balance=0, credit_limit=500):
        super().__init__(name, phone, package, balance)
        self.credit_limit = credit_limit

    def deduct(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be greater than 0.")

        if self.balance - amount < -self.credit_limit:
            raise Exception("Credit limit reached!")

        self.balance = self.balance - amount

    def show_debt(self):
        if self.balance < 0:
            print("Debt:", -self.balance, "EGP")
        else:
            print("No debt.")

    def __str__(self):
        return "[Postpaid] " + self.name + " | " + self.phone + " | " + self.package + " | Balance: " + str(self.balance) + " EGP | Limit: " + str(self.credit_limit) + " EGP"


subscribers = [
    PrepaidSubscriber("Ahmed Ali", "01501555555", "Bronze"),
    PrepaidSubscriber("Mona Hassan", "01512118555", "Silver"),
    PostpaidSubscriber("Karim Sami", "01525556789", "Gold"),
    PostpaidSubscriber("Sara Khaled", "01534555090", "Bronze"),
    Subscriber("Omar Mahmoud", "01645678901", "Silver", 50)
]


print("--- All Subscribers ---")
for s in subscribers:
    print(s)


def find_by_phone(subscribers, phone):
    for s in subscribers:
        if s.phone == phone:
            return s

    raise ValueError("Subscriber not found.")


def filter_by_package(subscribers, package):
    result = []

    for s in subscribers:
        if s.package == package:
            result.append(s)

    return result


def top_balance(subscribers):
    top = subscribers[0]

    for s in subscribers:
        if s.balance > top.balance:
            top = s

    return top

print("\nRecharging Ahmed Ali with 100 EGP...")
find_by_phone(subscribers, "01501555555").recharge(100)

print("\nUpgrading Ahmed Ali to Silver...")
find_by_phone(subscribers, "01501555555").upgrade_package("Silver")

print("\nChecking Ahmed's Data...")
find_by_phone(subscribers, "01501555555").check_data()

print("\nDeducting from Karim Sami...")
try:
    find_by_phone(subscribers, "01525556789").deduct(600)
except Exception as e:
    print(e)

print("\n--- Gold Subscribers ---")
for s in filter_by_package(subscribers, "Gold"):
    print(s)

print("\nTop Balance:")
print(top_balance(subscribers))
````
